```python
import numpy as np
import pandas as pd
import os
import re
import csv
from math import floor
from scipy.fft import fft
from sklearn.model_selection import KFold
from sklearn.preprocessing import StandardScaler
from sklearn.utils.class_weight import compute_class_weight
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score,
    roc_auc_score, average_precision_score
)
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping
import matplotlib.pyplot as plt
import cv2
from google.colab.patches import cv2_imshow
from google.colab import drive

drive.mount('/content/drive')

# =====================
# 配置参数
# =====================
WINDOW_SIZE = 256   # 窗口大小（帧）
FPS = 30            # 帧率
STEP = 1            # 滑动窗口步长
THRESHOLD = 0.2     # 投票阈值
MODALITIES = ['flow', 'volume']  # 使用的模态

# 新数据根目录
BASE_DIR = "/content/drive/MyDrive/New Data needs"

# 仅处理这些 test_number
TEST_NUMBERS = ["1", "2", "2-2", "3", "12", "12-2", "26"]

# 是否进行交互式输入起止时间（Colab/本地均可用）
INTERACTIVE_TIME_INPUT = True

# =====================
# 工具函数
# =====================
def parse_time_to_seconds(s: str):
    """
    支持 'HH:MM:SS' / 'MM:SS' / 'SS' 三种格式。
    返回秒(int)。若 s 为空字符串或 None，返回 None。
    """
    if s is None:
        return None
    s = str(s).strip()
    if s == "":
        return None
    parts = s.split(':')
    try:
        if len(parts) == 3:
            h, m, sec = map(int, parts)
            return h * 3600 + m * 60 + sec
        elif len(parts) == 2:
            m, sec = map(int, parts)
            return m * 60 + sec
        elif len(parts) == 1:
            return int(parts[0])
        else:
            return None
    except Exception:
        return None


def prompt_time_range(test_number: str, default_start_sec: int):
    """
    交互式输入：每个 test_number 的起止时间。
    回车：起始=默认(.npy首帧时间)；结束=到结尾。
    """
    print(f"\n--- 为测试 {test_number} 选择 CSV/NPY 保留区间 ---")
    print(f"默认起始时间（自动与 .npy 对齐）：{default_start_sec} 秒（约 {default_start_sec//3600:02d}:{(default_start_sec%3600)//60:02d}:{default_start_sec%60:02d}）")
    start_in = input("请输入起始时间 (HH:MM:SS 或 MM:SS 或 SS)，直接回车=默认：").strip()
    end_in = input("请输入结束时间 (HH:MM:SS 或 MM:SS 或 SS)，直接回车=不裁剪到结尾：").strip()

    start_sec = parse_time_to_seconds(start_in)
    end_sec = parse_time_to_seconds(end_in)

    if start_sec is None:
        start_sec = default_start_sec

    if end_sec is not None and end_sec < start_sec:
        print("⚠️ 结束时间早于起始时间，忽略结束时间（按到结尾处理）。")
        end_sec = None

    return start_sec, end_sec


# =====================
# 1. 数据读取
# =====================
def load_data(test_number: str):
    """
    直接按给定的 test_number 从 BASE_DIR 读取 .npy 和 .csv
    - flow_data_{test_number}.npy
    - volume_data_{test_number}.npy
    - d_smoothed_{test_number}.npy
    - frame_times_{test_number}.npy
    - dataset_{test_number}.csv
    """
    def _p(fname): return os.path.join(BASE_DIR, fname)

    flow = np.load(_p(f"flow_data_{test_number}.npy"))
    volume = np.load(_p(f"volume_data_{test_number}.npy"))
    d_smoothed = np.load(_p(f"d_smoothed_{test_number}.npy"))
    frame_times = np.load(_p(f"frame_times_{test_number}.npy"))
    annot_path = _p(f"dataset_{test_number}.csv")
    annot = pd.read_csv(annot_path)

    return {
        'flow': flow,
        'volume': volume,
        'd_smoothed': d_smoothed,
        'frame_times': frame_times,
        'annot': annot,
        'test_number': test_number
    }


# =====================
# 2. 从 CSV 构建“对齐到 .npy 起点”的帧标签（支持再按裁剪偏移对齐）
# =====================
def build_frame_labels(
    annot: pd.DataFrame,
    fps: int,
    frame_times: np.ndarray,
    start_time_str: str = None,
    end_time_str: str = None,
    local_offset_frames: int = 0
):
    """
    核心：把 CSV 的“全局帧索引”转换成 **对齐到 .npy 起点** 的“局部帧索引”，
    再整体左移 local_offset_frames，使标签与“被裁剪后的 .npy 数据（从0开始）”对齐。
    """
    npy_start_offset_frames = int(floor(float(frame_times[0]) * fps))

    user_start_sec = parse_time_to_seconds(start_time_str)
    user_end_sec = parse_time_to_seconds(end_time_str)
    if user_start_sec is None:
        user_start_sec = int(floor(float(frame_times[0])))

    start_global_frame = int(user_start_sec * fps)
    end_global_frame = None
    if user_end_sec is not None:
        end_global_frame = int(user_end_sec * fps + (fps - 1))

    frame_labels = {}
    for _, row in annot.iterrows():
        tsec = parse_time_to_seconds(str(row['time']))
        if tsec is None:
            continue
        try:
            frm = int(row['frame'])
        except Exception:
            continue

        global_frame = int(tsec * fps + frm - 1)

        if global_frame < start_global_frame:
            continue
        if end_global_frame is not None and global_frame > end_global_frame:
            continue

        local_frame = global_frame - npy_start_offset_frames
        if local_frame < 0:
            continue

        local_frame_after_crop = local_frame - local_offset_frames
        if local_frame_after_crop < 0:
            continue

        lab = str(row['label']).strip().lower()
        lbl = 1 if lab == 'abnormal' else 0
        frame_labels[local_frame_after_crop] = lbl

    return frame_labels


# =====================
# 2.5 对 .npy 数据做时间窗裁剪
# =====================
def crop_data_by_time_range(
    data: dict,
    fps: int,
    start_time_str: str = None,
    end_time_str: str = None
):
    frame_times = data['frame_times']
    n_frames = len(frame_times)
    if n_frames == 0:
        raise ValueError("frame_times 为空，无法裁剪。")

    npy_start_offset_frames = int(floor(float(frame_times[0]) * fps))

    user_start_sec = parse_time_to_seconds(start_time_str)
    user_end_sec = parse_time_to_seconds(end_time_str)
    if user_start_sec is None:
        user_start_sec = int(floor(float(frame_times[0])))

    start_global_frame = int(user_start_sec * fps)
    end_global_frame = None
    if user_end_sec is not None:
        end_global_frame = int(user_end_sec * fps + (fps - 1))

    crop_start_local = max(0, start_global_frame - npy_start_offset_frames)
    if end_global_frame is not None:
        crop_end_local_inclusive = min(n_frames - 1, end_global_frame - npy_start_offset_frames)
    else:
        crop_end_local_inclusive = n_frames - 1

    if crop_end_local_inclusive < crop_start_local:
        raise ValueError(
            f"裁剪时间段为空（start={user_start_sec}s, end={user_end_sec if user_end_sec is not None else 'END'}s）。"
        )

    s = crop_start_local
    e = crop_end_local_inclusive + 1

    cropped = {
        'flow': data['flow'][s:e],
        'volume': data['volume'][s:e],
        'd_smoothed': data['d_smoothed'][s:e],
        'frame_times': data['frame_times'][s:e],
        'annot': data['annot'],
        'test_number': data['test_number']
    }

    return cropped, crop_start_local, crop_end_local_inclusive


# =====================
# 3. 滑动窗口
# =====================
def create_sliding_windows(data, window_size=WINDOW_SIZE, step=STEP):
    windows = []
    start_indices = []

    for modality in MODALITIES:
        signal = data[modality]
        num_frames = len(signal)
        modality_windows = []

        for start in range(0, num_frames - window_size + 1, step):
            end = start + window_size
            window = signal[start:end]
            modality_windows.append(window)
            if modality == MODALITIES[0]:
                start_indices.append(start)

        windows.append(modality_windows)

    combined_windows = []
    for i in range(len(windows[0])):
        combined_window = []
        for j in range(len(MODALITIES)):
            combined_window.append(windows[j][i])
        combined_windows.append(np.concatenate(combined_window))

    return combined_windows, start_indices


# =====================
# 4. 窗口标签
# =====================
def generate_labels_from_frame_labels(frame_labels: dict, start_indices, window_size=WINDOW_SIZE):
    labels = []
    for start_idx in start_indices:
        window_labels = []
        for i in range(start_idx, start_idx + window_size):
            if i in frame_labels:
                window_labels.append(frame_labels[i])
            else:
                window_labels.append(0)
        abnormal_ratio = np.mean(window_labels)
        if abnormal_ratio >= THRESHOLD:
            labels.append(1)
        elif abnormal_ratio <= (1 - THRESHOLD):
            labels.append(0)
        else:
            labels.append(-1)
    return np.array(labels)


# =====================
# 5. FFT 特征
# =====================
def extract_fft_features(windows):
    fft_features = []
    for window in windows:
        fft_result = np.abs(fft(window))
        fft_features.append(fft_result[:len(fft_result)//2])
    return np.array(fft_features)


# =====================
# 6. 单个样本管道
# =====================
def process_one(data, start_time_str=None, end_time_str=None):
    cropped_data, crop_start_local, crop_end_local_inclusive = crop_data_by_time_range(
        data,
        fps=FPS,
        start_time_str=start_time_str,
        end_time_str=end_time_str
    )

    windows, start_indices = create_sliding_windows(cropped_data)

    frame_labels = build_frame_labels(
        annot=cropped_data['annot'],
        fps=FPS,
        frame_times=data['frame_times'],
        start_time_str=start_time_str,
        end_time_str=end_time_str,
        local_offset_frames=crop_start_local
    )

    labels = generate_labels_from_frame_labels(
        frame_labels,
        start_indices,
        window_size=WINDOW_SIZE
    )

    features = extract_fft_features(windows)

    valid_mask = labels != -1
    features = features[valid_mask]
    labels = labels[valid_mask]

    return features, labels, data['test_number']


# =====================
# 7. 主训练流程（新增：手动选择保留折）
# =====================
def main():
    all_features = []
    all_labels = []
    video_ids = []

    print("开始处理指定的测试数据...")
    for tn in TEST_NUMBERS:
        try:
            data = load_data(tn)
            default_start_sec = int(floor(float(data['frame_times'][0])))

            if INTERACTIVE_TIME_INPUT:
                start_sec, end_sec = prompt_time_range(tn, default_start_sec)
                start_str = None if start_sec is None else f"{start_sec}"
                end_str = None if end_sec is None else f"{end_sec}"
            else:
                start_str, end_str = str(default_start_sec), None

            features, labels, test_number = process_one(
                data,
                start_time_str=start_str,
                end_time_str=end_str
            )

            all_features.append(features)
            all_labels.append(labels)
            video_ids.extend([test_number] * len(labels))

            kept = f"{start_str or default_start_sec} sec -> {end_str if end_str is not None else 'END'}"
            print(f"处理测试 {test_number} 完成, 样本数: {len(labels)}，保留区间: {kept}")

        except FileNotFoundError as e:
            print(f"⚠️ 跳过 {tn}：文件缺失 -> {e}")
        except Exception as e:
            print(f"⚠️ 跳过 {tn}：处理时发生异常 -> {e}")

    if len(all_features) == 0:
        raise RuntimeError("未成功加载到任何数据，请检查文件路径与文件名是否正确。")

    X = np.concatenate(all_features)
    y = np.concatenate(all_labels)
    video_ids = np.array(video_ids)

    print(f"\n总样本数: {len(X)}")
    print(f"类别分布 - 正常: {np.sum(y == 0)}, 异常: {np.sum(y == 1)}")

    unique_video_ids = np.unique(video_ids)
    if len(unique_video_ids) < 2:
        raise RuntimeError("有效的 test_number 数量不足以进行交叉验证。")
    n_splits = min(5, len(unique_video_ids))
    kf = KFold(n_splits=n_splits, shuffle=True, random_state=42)

    # 新增：每折详细记录
    per_fold_stats = []

    results = {
        'accuracy': [], 'precision': [], 'recall': [],
        'f1': [], 'roc_auc': [], 'pr_auc': []
    }

    fold = 1
    for train_idx, test_idx in kf.split(unique_video_ids):
        print(f"\n=== 正在处理 Fold {fold} / {n_splits} ===")
        train_videos = unique_video_ids[train_idx]
        test_videos = unique_video_ids[test_idx]

        train_mask = np.isin(video_ids, train_videos)
        test_mask = np.isin(video_ids, test_videos)
        X_train, X_test = X[train_mask], X[test_mask]
        y_train, y_test = y[train_mask], y[test_mask]

        print(f"训练集大小: {len(X_train)}, 测试集大小: {len(X_test)}")
        print(f"训练集类别分布 - 正常: {np.sum(y_train == 0)}, 异常: {np.sum(y_train == 1)}")
        print(f"测试集类别分布 - 正常: {np.sum(y_test == 0)}, 异常: {np.sum(y_test == 1)}")

        # 记录是否“测试集单类”
        test_unique = np.unique(y_test)
        degenerate = (len(test_unique) == 1)

        scaler = StandardScaler()
        X_train_scaled = scaler.fit_transform(X_train)
        X_test_scaled = scaler.transform(X_test)

        try:
            classes = np.unique(y_train)
            class_weights = compute_class_weight(class_weight='balanced', classes=classes, y=y_train)
            class_weight_dict = {int(c): w for c, w in zip(classes, class_weights)}
        except Exception as e:
            print(f"⚠️ 计算类别权重失败({e})，使用等权重。")
            class_weight_dict = {0: 1.0, 1: 1.0}

        input_dim = X_train_scaled.shape[1]
        model = Sequential([
            Dense(128, activation='relu', input_shape=(input_dim,)),
            Dropout(0.3),
            Dense(64, activation='relu'),
            Dropout(0.3),
            Dense(1, activation='sigmoid')
        ])
        model.compile(optimizer=Adam(learning_rate=0.001), loss='binary_crossentropy', metrics=['accuracy'])

        early_stopping = EarlyStopping(
            monitor='val_loss',
            patience=5,
            restore_best_weights=True
        )

        history = model.fit(
            X_train_scaled, y_train,
            epochs=50, batch_size=256,
            validation_split=0.2,
            class_weight=class_weight_dict,
            callbacks=[early_stopping],
            verbose=1
        )

        y_pred = model.predict(X_test_scaled)
        y_pred_binary = (y_pred > 0.5).astype(int)

        accuracy = accuracy_score(y_test, y_pred_binary)
        precision = precision_score(y_test, y_pred_binary, zero_division=0)
        recall = recall_score(y_test, y_pred_binary, zero_division=0)
        f1 = f1_score(y_test, y_pred_binary, zero_division=0)
        try:
            roc_auc = roc_auc_score(y_test, y_pred)
        except Exception:
            roc_auc = np.nan
        try:
            pr_auc = average_precision_score(y_test, y_pred)
        except Exception:
            pr_auc = np.nan

        # 保存到总表
        results['accuracy'].append(accuracy)
        results['precision'].append(precision)
        results['recall'].append(recall)
        results['f1'].append(f1)
        results['roc_auc'].append(roc_auc)
        results['pr_auc'].append(pr_auc)

        print(f"\nFold {fold} 评估结果:")
        print(f"Accuracy: {accuracy:.4f}")
        print(f"Precision: {precision:.4f}")
        print(f"Recall: {recall:.4f}")
        print(f"F1 Score: {f1:.4f}")
        print(f"ROC AUC: {roc_auc if not np.isnan(roc_auc) else 'NA'}")
        print(f"PR AUC: {pr_auc if not np.isnan(pr_auc) else 'NA'}")
        if degenerate:
            print("⚠️ 提示：该折测试集为单一类别（Degenerate），部分指标（如 AUC）可能不可计算。")

        # 记录该折的全部信息，供后续“人工选择保留折”使用
        per_fold_stats.append({
            'fold': fold,
            'degenerate': bool(degenerate),
            'n_test': int(len(y_test)),
            'test_pos': int(np.sum(y_test == 1)),
            'test_neg': int(np.sum(y_test == 0)),
            'accuracy': float(accuracy),
            'precision': float(precision),
            'recall': float(recall),
            'f1': float(f1),
            'roc_auc': float(roc_auc) if not np.isnan(roc_auc) else np.nan,
            'pr_auc': float(pr_auc) if not np.isnan(pr_auc) else np.nan
        })

        # 可视化训练历史
        plt.figure(figsize=(12, 5))
        plt.subplot(1, 2, 1)
        plt.plot(history.history['accuracy'], label='训练准确率')
        plt.plot(history.history['val_accuracy'], label='验证准确率')
        plt.title(f'Fold {fold} - 准确率')
        plt.ylabel('准确率')
        plt.xlabel('Epoch')
        plt.legend()

        plt.subplot(1, 2, 2)
        plt.plot(history.history['loss'], label='训练损失')
        plt.plot(history.history['val_loss'], label='验证损失')
        plt.title(f'Fold {fold} - 损失')
        plt.ylabel('损失')
        plt.xlabel('Epoch')
        plt.legend()
        plt.tight_layout()
        plt.show()

        fold += 1

    # 展示每折摘要，便于人工选择
    print("\n=== 每折摘要（供选择保留折用） ===")
    df_folds = pd.DataFrame(per_fold_stats)
    # 打印时将 NaN 以 NA 显示
    with pd.option_context('display.max_columns', None, 'display.width', 120):
        print(df_folds.fillna('NA'))

    # 让用户输入要保留的折 ID（英文逗号，无空格）；直接回车=保留全部
    keep_input = input("\n请输入要保留参与最终统计的折ID（英文逗号、无空格；直接回车=保留全部）：").strip()
    if keep_input == "":
        kept_folds = set(df_folds['fold'].tolist())
    else:
        try:
            kept_folds = set(int(x) for x in keep_input.split(',') if x != "")
        except Exception:
            print("⚠️ 输入解析失败，默认保留全部折。")
            kept_folds = set(df_folds['fold'].tolist())

    # 过滤出被保留的折
    kept_mask = df_folds['fold'].isin(sorted(kept_folds))
    kept_df = df_folds[kept_mask].copy()
    if kept_df.empty:
        print("⚠️ 未选择到任何折，默认使用全部折。")
        kept_df = df_folds.copy()

    # 最终评估结果（对每个指标：按可用数值计算 mean ± std，NaN 自动跳过）
    print("\n=== 最终评估结果（基于保留折） ===")
    metrics = ['accuracy', 'precision', 'recall', 'f1', 'roc_auc', 'pr_auc']
    final_summary = {}
    for m in metrics:
        vals = kept_df[m].to_numpy(dtype=float)
        mean_val = np.nanmean(vals)
        std_val = np.nanstd(vals)
        final_summary[m] = (mean_val, std_val)
        print(f"{m.capitalize()}: {mean_val:.4f} ± {std_val:.4f} (基于 {np.sum(~np.isnan(vals))} 个有效折)")

    # 保存每折与最终结果
    fold_stats_path = os.path.join(BASE_DIR, "mlp_cv_fold_stats.csv")
    kept_df.to_csv(fold_stats_path, index=False)

    result_df = pd.DataFrame({
        'metric': metrics,
        'mean': [final_summary[m][0] for m in metrics],
        'std': [final_summary[m][1] for m in metrics],
        'kept_folds': [",".join(str(f) for f in sorted(kept_folds))]*len(metrics)
    })
    result_path = os.path.join(BASE_DIR, "mlp_cross_validation_results.csv")
    result_df.to_csv(result_path, index=False)

    print(f"\n✅ 训练完成！\n已保存每折统计到: {fold_stats_path}\n已保存最终汇总到: {result_path}")
    print(f"最终保留的折: {','.join(str(f) for f in sorted(kept_folds))}")

if __name__ == "__main__":
    main()
```
